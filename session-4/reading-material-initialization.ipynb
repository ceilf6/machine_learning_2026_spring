```python
import torch
import torch.nn as nn


def xavier_uniform_(tensor, gain=1.0):
    """Xavier/Glorot initialization."""
    fan_in, fan_out = tensor.size(0), tensor.size(1)
    limit = gain * torch.sqrt(torch.tensor(6.0 / (fan_in + fan_out)))
    return tensor.uniform_(-limit, limit)


# PyTorch built-in
layer = nn.Linear(512, 512)
nn.init.xavier_uniform_(layer.weight)


# Print full weight matrix
print("First layer weights:")
print(layer.weight)

# Or just inspect a small part (recommended)
print("\nFirst 5x5 block of weights:")
print(layer.weight[:5, :5])
```

```python
def kaiming_normal_(tensor, mode="fan_in", nonlinearity="relu"):
    """Kaiming/He initialization."""
    fan = tensor.size(1) if mode == "fan_in" else tensor.size(0)

    # Compute gain based on nonlinearity
    if nonlinearity == "relu":
        gain = torch.sqrt(torch.tensor(2.0))
    elif nonlinearity == "linear":
        gain = 1.0
    else:
        raise ValueError(f"Unsupported nonlinearity: {nonlinearity}")

    std = gain / torch.sqrt(torch.tensor(fan))
    return tensor.normal_(0, std)


# PyTorch built-in
layer = nn.Linear(512, 512)
nn.init.kaiming_normal_(layer.weight, nonlinearity="relu")

# Print full weight matrix
print("First layer weights:")
print(layer.weight)

# Or just inspect a small part (recommended)
print("\nFirst 5x5 block of weights:")
print(layer.weight[:5, :5])
```
